11 Identify and remove outliers in the km_driven column using the IQR methodB

```python
#outliers
outliers = df[(df["km_driven"] < lower_bound) | (df["km_driven"] > upper_bound)]
```

```python
outliers
```

```python
Q1 = df["km_driven"].quantile(0.25)
Q3 = df["km_driven"].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df_cleaned = df[(df["km_driven"] >= lower_bound) & (df["km_driven"] <= upper_bound)]
print(f"Outliers removed: {len(df) - len(df_cleaned)}")
```

```python
plt.figure(figsize = (10,6))
sns.boxplot(x = "km_driven", data  = df, palette = "Set1") 
plt.show()
```

9. Does the duration of tracks vary significantly across different artists? Explore this visually using a box plot or
violin plot?

```python
import scipy.stats as stats
plt.figure(figsize=(30, 20))
sns.boxplot(x="Artist", y="Duration (ms)", data=df)
plt.xticks(rotation=90)
plt.xlabel("Artist")
plt.ylabel("Duration (ms)")
plt.title("Duration of Tracks Across Different Artists (Box Plot)")
plt.show()

plt.figure(figsize=(30, 20))
sns.violinplot(x="Artist", y="Duration (ms)", data=df)
plt.xticks(rotation=90)
plt.xlabel("Artist")
plt.ylabel("Duration (ms)")
plt.title("Duration of Tracks Across Different Artists (Violin Plot)")
plt.show()


artist_groups = [group["Duration (ms)"].values for name, group in df.groupby("Artist")]
stat, p_value = stats.f_oneway(*artist_groups)

print("ANOVA Test Results:")
print(f"F-statistic: {stat:.4f}")
print(f"p-value: {p_value:.4f}")

if p_value < 0.05:
    print("There is a significant variation in track durations across different artists.")
else:
    print("There is no significant variation in track durations across different artists.")
```
